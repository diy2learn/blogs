```python
# Import libs
from keras.preprocessing import image
from keras.layers import Conv2D, GlobalAveragePooling2D, Dropout, Flatten, Dense
from keras.models import Model
from keras.callbacks import LearningRateScheduler, ModelCheckpoint
from keras.callbacks import ReduceLROnPlateau, EarlyStopping
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from keras.optimizers import SGD
```

```python
from keras.applications import xception
```

```python
from keras.models import load_model
```

```python
#from keras.applications import vgg16
from keras.callbacks import TensorBoard
```

```python
import pickle
```

```python
import pydot
```

```python
# Pars
train_dir = "./data/train"
val_dir = "./data/val"
INPUT_SIZE = 499
#BATCH_SIZE = 64 # For transfer learning; out of memory when fine tune
BATCH_SIZE = 32 # reduce from 64 to avoid out of memory error
MODEL_NAME = "xception"
DESCRIPTION = "FineTune05"
NB_LAYERS_TO_FREEZE = 126
EPOCHS_tl = 30
EPOCHS_ft = 50
INITIAL_EPOCHS = 0
FC_SIZE = 1024
N_CLASSES = 120
preprocessing_function = xception.preprocess_input
```

```python
# Data augmentation 
train_gen = image.ImageDataGenerator(featurewise_center=False,
                                     samplewise_center=False,
                                     featurewise_std_normalization=False,
                                     samplewise_std_normalization=False,
                                     zca_whitening=False,
                                     rotation_range=15.0,
                                     width_shift_range=0.125,
                                     height_shift_range=0.125,
                                     #shear_range=0.0,
                                     zoom_range=0.2,
                                     channel_shift_range=0.0,
                                     fill_mode='nearest',
                                     cval=0.0,
                                     horizontal_flip=True,
                                     vertical_flip=False,
                                     rescale=None,
                                     preprocessing_function=preprocessing_function,
                                     data_format=None)
val_gen = image.ImageDataGenerator(preprocessing_function=preprocessing_function)
```

```python
# Need find way to do this if flow_from_directory:
#train_gen.fit(x_train)
#val_gen.fit(x_train)# 
# Data augmentation
train_generator = train_gen.flow_from_directory(train_dir,target_size=(INPUT_SIZE,INPUT_SIZE),batch_size=BATCH_SIZE)
val_generator = val_gen.flow_from_directory(val_dir,target_size=(INPUT_SIZE,INPUT_SIZE),batch_size=BATCH_SIZE)
```

```python
# base_model
base_model = xception.Xception(weights='imagenet',include_top=False)
# If we chose vgg16 for base_model
#base_model = vgg16.VGG16(include_top=False, weights='imagenet')
```

```python
#base_model.summary()
# Plot base_model schema
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
SVG(model_to_dot(base_model).create(prog='dot',format='svg'))
```

```python
# Let visualize name of each layer to decide which layer to freeze training
# if we refine tune only last block, then it starts from 126
for i, layer in enumerate(base_model.layers):
    print(i, layer.name)
```

```python
# Add dense layer on top
def add_top_layer(base_model,n_classes):
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.1)(x)
    x = Dense(FC_SIZE, activation='relu')(x)
    x = Dropout(0.5)(x)
    predictions = Dense(n_classes, init='glorot_uniform',activation='softmax')(x)
    model = Model(inputs=base_model.input,outputs=predictions,name="xception_neck_FC")
    return model
```

```python
# Setup training
save_dir = os.path.join(os.getcwd(), 'saved_models')
```

```python
# Setup transfer learning
# Prepare model model saving directory.
#model_name = '%s-%s-%s-{epoch:03d}-{val_acc:.3f}-{acc:.3f}.h5' %(MODEL_NAME,DESCRIPTION,"transferLearn")
model_name = 'best-%s-%s-%s.h5' %(MODEL_NAME,DESCRIPTION,"transferLearn")
if not os.path.isdir(save_dir):
    os.makedirs(save_dir)
filepath_tf = os.path.join(save_dir, model_name)

# Prepare callbacks for model saving and for learning rate adjustment.
checkpoint = ModelCheckpoint(filepath=filepath_tf,
                             monitor='val_acc',
                             verbose=1,
                             save_best_only=True)

ts_graph_dir = os.path.join(save_dir,'ts_Graph_%s_%s_%s'%(MODEL_NAME,DESCRIPTION,"transferLearn"))
if not os.path.isdir(ts_graph_dir):
    os.makedirs(ts_graph_dir)
ts_board = TensorBoard(log_dir= ts_graph_dir, histogram_freq=0,
                          write_graph=True, write_images=False)
early_stop = EarlyStopping(monitor='val_loss', patience=10)

callbacks_tl = [checkpoint,ts_board,early_stop]
def setup_transfer_learn(base_model,model):
    
    for layer in base_model.layers:
        layer.trainable = False
    opt = 'rmsprop'
    model.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# Setup for fine tune
def lr_schedule(epoch):
    """Learning Rate Schedule
    Learning rate is scheduled to be reduced after 5, 30, 50, 80 epochs.
    Called automatically every epoch as part of callbacks during training.
    # Arguments
        epoch (int): The number of epochs
    # Returns
        lr (float32): learning rate
    """
    lr = 1e-4
#     if epoch > 5:
#         lr *= 0.5
#     elif epoch > 30:
#         lr *= 0.5
#     elif epoch > 50:
#         lr *= 0.5
#     print('Learning rate: ', lr)
    return lr
# Learning rate
lr_scheduler = LearningRateScheduler(lr_schedule)
# Prepare model model saving directory.

#model_name = '%s-%s-{epoch:03d}-{val_acc:.3f}-{acc:.3f}.h5' %(MODEL_NAME,DESCRIPTION)
# Save only the best
model_name = 'best-%-s-%s-%s.h5' %(MODEL_NAME,DESCRIPTION,"fineTune")
if not os.path.isdir(save_dir):
    os.makedirs(save_dir)
filepath = os.path.join(save_dir, model_name)

# Prepare callbacks for model saving and for learning rate adjustment.
checkpoint = ModelCheckpoint(filepath=filepath,
                             monitor='val_acc',
                             verbose=1,
                             save_best_only=True)


lr_reducer = ReduceLROnPlateau(factor=np.sqrt(0.1),
                               cooldown=0,
                               patience=5,
                               min_lr=0.5e-6)
ts_graph_dir = os.path.join(save_dir,'ts_Graph_%s_%s_%s'%(MODEL_NAME,DESCRIPTION,"fineTune"))
if not os.path.isdir(ts_graph_dir):
    os.makedirs(ts_graph_dir)
ts_board = TensorBoard(log_dir= ts_graph_dir, histogram_freq=0,
                          write_graph=True, write_images=False)

early_stop = EarlyStopping(monitor='val_loss', patience=20)
callbacks_ft = [checkpoint, ts_board]


def setup_fine_tune(model):
    # NB_LAYERS_TO_FREEZE points to the layer in your model you want to train.
    # For example if you want to train the last block of a 19 layer VGG16 model this should be 15
    # If you want to train the last Two blocks of an Inception model it should be 172
    # layers before this number will used the pre-trained weights, layers above and including this number
    # will be re-trained based on the new data.
    for layer in model.layers[:NB_LAYERS_TO_FREEZE]:
        layer.trainable = False
    for layer in model.layers[NB_LAYERS_TO_FREEZE:]:
        layer.trainable = True
    opt = SGD(lr=1e-4, momentum=.9)    
    model.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])
        
```

```python
# Plot summary of training
def plot_train_history(history):
    loss_all = history.history['loss']
    acc_all = history.history['acc']
    val_loss_all = history.history['val_loss']
    val_acc_all = history.history['val_acc']
    plt.close('all')
    fig, (ax1, ax2) = plt.subplots(1, 2, sharey=False)
    ax1.plot(loss_all)
    ax1.plot(val_loss_all)
    ax1.set_xlabel('epoch')
    ax1.set_ylabel('loss')
    ax1.legend(['loss','val_loss'],loc='upper right')
    #plt.show()
    # Plot accuracy
    ax2.plot(acc_all)
    ax2.plot(val_acc_all)
    ax2.set_xlabel('epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend(['Acc','Val_Acc'],loc='lower right')
    plt.tight_layout()
```

```python
# TRANSFER LEARNING
model = add_top_layer(base_model,N_CLASSES)
setup_transfer_learn(base_model,model)
# Training
history1 = model.fit_generator(train_generator,validation_data=val_generator,epochs=EPOCHS_tl,
                             initial_epoch = INITIAL_EPOCHS, workers=12,callbacks=callbacks_tl)
```
